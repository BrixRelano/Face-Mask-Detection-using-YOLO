```python
#Importing necessary libraries and instantiating numpy
import cv2
import numpy as np

#Reading training weights file and testing files for referencing
net = cv2.dnn.readNet('C:\\Users\\PC\\Downloads\\yolov3_training_last.weights', 'C:\\Users\\PC\\Downloads\\yolov3_testing.cfg')

#Declaring necessary variables
label = ' '
classes = []

#Opening classes.txt file containing the two classes used
with open("C:\\Users\\PC\\Downloads\\classes.txt", "r") as f:
    classes = f.read().splitlines()

#Declaring variables for video capture, font, and colors
cap = cv2.VideoCapture(0)
font = cv2.FONT_HERSHEY_PLAIN
colors = np.random.uniform(0, 255, size=(100, 3))

#While loop to read the video capture frame per frame
while True:
    _, img = cap.read()
    height, width, _ = img.shape

#Transforming the frame into BNW and processing it for easier analysis
    blob = cv2.dnn.blobFromImage(img, 1/255, (416, 416), (0,0,0), swapRB=True, crop=False)
    net.setInput(blob)
    output_layers_names = net.getUnconnectedOutLayersNames()
    layerOutputs = net.forward(output_layers_names)

    boxes = []
    confidences = []
    class_ids = []

#Calculating the dimensions of the bounding boxes, as well as the confidence level of the detection
    for output in layerOutputs:
        for detection in output:
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]
            if confidence > 0.2:
                center_x = int(detection[0]*width)
                center_y = int(detection[1]*height)
                w = int(detection[2]*width)
                h = int(detection[3]*height)

                x = int(center_x - w/2)
                y = int(center_y - h/2)

                boxes.append([x, y, w, h])
                confidences.append((float(confidence)))
                class_ids.append(class_id)

    indexes = cv2.dnn.NMSBoxes(boxes, confidences, 0.2, 0.4)

#Applying the bounding boxes, labels, and confidence per detection
    if len(indexes)>0:
        for i in indexes.flatten():
            x, y, w, h = boxes[i]
            label = str(classes[class_ids[i]])
            confidence = str(round(confidences[i],2))
            color = colors[i]
            cv2.rectangle(img, (x,y), (x+w, y+h), color, 2)
            cv2.putText(img, label + " " + confidence, (x, y+20), font, 2, (255,255,255), 2)

#Showing the Live webcam feed           
    cv2.imshow('Webcam Feed', img)
    key = cv2.waitKey(1)
#If the 'Esc' key is pressed, the live feed and the program will terminate
    if key==27:
        break

cap.release()
cv2.destroyAllWindows()
```
